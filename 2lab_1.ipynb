```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
# Завантаження та очищення даних
df = pd.read_csv('FIFA_World_Cup.csv').dropna()
```

```python
# Стандартизація числових даних
scaler = StandardScaler()
numeric_data = df.select_dtypes(include=[int, float])
scaled_data = scaler.fit_transform(numeric_data)
```

```python
# Використання PCA для зменшення розмірності
pca = PCA(n_components=2)
reduced_data_pca = pca.fit_transform(scaled_data)

# Використання t-SNE для зменшення розмірності
tsne = TSNE(n_components=2, random_state=42)
reduced_data_tsne = tsne.fit_transform(scaled_data)
```

```python
# Функція для візуалізації даних
def plot_data(data, title, xlabel, ylabel):
    plt.figure(figsize=(8, 6))
    plt.scatter(data[:, 0], data[:, 1])
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.colorbar()
    plt.show()
```

```python
# Візуалізація даних PCA
plot_data(reduced_data_pca, 'Графік розсіювання для зменшеного набору даних (PCA)', 'Перша компонента PCA', 'Друга компонента PCA')

# Візуалізація даних t-SNE
plot_data(reduced_data_tsne, 'Графік розсіювання для зменшеного набору даних (t-SNE)', 'Перша компонента t-SNE', 'Друга компонента t-SNE')
```
